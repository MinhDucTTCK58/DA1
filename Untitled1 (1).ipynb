```python
# Import library
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#draw a line
x=[1,2,3]
y=[2,3,4]
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#plot two lines on the same pot with legends
x1=[1,2,3]
y1=[2,3,4]
plt.plot(x1,y1,label="line 1")
x2=[3,4,5]
y2=[4,5,6]
plt.plot(x2,y2,label="line 2")
plt.legend()
plt.show()
```

```python
#draw scatter graph
from pylab import randn
x=randn(200)
y=randn(200)
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter graph")
plt.show()
```
